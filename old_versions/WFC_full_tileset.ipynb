```python
import os
import sys
if "newtiles" in sys.modules.keys():
    del sys.modules["newtiles"]
import newtiles as t
import pygame
from typing import List
import numpy as np

try:
    pygame.quit()
except:
    pass

boardSize = 8
screensize = 1000
tilesize = screensize // boardSize



pygame.init()
pygame.display.set_caption('Tiles Demo')
screen = pygame.display.set_mode((screensize, screensize))
pygame.event.pump()
pygame.display.update()
pygame.time.wait(100)

import pandas as pd

tileFolder = "map_tiles"
tileData = pd.read_csv(os.path.join(tileFolder, "tile_data.csv"))


def imageTo4Tiles(surface : pygame.surface, pattern : List):
    tile : t.ImageTile = t.ImageTile(pattern, surface)
    all4 : List[t.Tile] = [tile, tile.copy().rot90(), tile.copy().rot90().rot90(), tile.copy().rot90().rot90().rot90()]
    to_be_removed = [False] * 4
    for t1 in range(4):
        if to_be_removed[t1]: continue

        for t2 in range(t1+1, 4):
            to_be_removed[t2] |= all4[t1].sidePatterns == all4[t2].sidePatterns 
    
    return [all4[i] for i in range(4) if not to_be_removed[i]]
            




alltiles = []

for i in range(len(tileData)):
    pattern = list(tileData.iloc[i]["pattern"])
    surface = pygame.image.load(os.path.join(tileFolder, tileData.iloc[i]['location'])).convert()   
    # surface = pygame.transform.scale2x(pygame.transform.scale2x(surface)) 
    
    alltiles.extend(imageTo4Tiles(surface, pattern))



# initializing the manifold with all tiles possible

manifold = t.Manifold(boardSize, boardSize)
for x in range(boardSize):
    for y in range(boardSize):
        manifold.collections.append(t.TileCollection(alltiles))

# manifold.collections[y*boardSize + x].tiles = [straight,]


def showBoard(speed_ms : int = 100):
    
    screen.fill((100, 100, 100))

        

    # draw all the tiles
    for x in range(boardSize):
        for y in range(boardSize):
            
            manifold.collections[y * boardSize + x].draw(screen, np.array([x, y]), tilesize)
            pygame.time.wait(speed_ms // (boardSize * boardSize))

            if speed_ms > 0:
                pygame.event.pump()
                pygame.display.update()
    
    # draw dual grid
    for x in range(0, boardSize):
        pygame.draw.line(screen, (100, 100, 200), ((x+0.5) * tilesize, 0), ((x+0.5) * tilesize, screensize), width=5)
        pygame.draw.line(screen, (100, 100, 200), (0, (x+0.5) * tilesize), (screensize, (x+0.5) * tilesize), width=5)

        # draw normal grid
    for x in range(1, boardSize):
        pygame.draw.line(screen, (0, 0, 0), (x * tilesize, 0), (x * tilesize, screensize), width=1)
        pygame.draw.line(screen, (0, 0, 0), (0, x * tilesize), (screensize, x * tilesize), width=1)
    
    pygame.event.pump()
    pygame.display.update()

def delay(ms : int):
    pygame.time.wait(ms)
    pygame.event.pump()
    pygame.display.update()

showBoard(speed_ms=200)


while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                pygame.quit()
            if event.key == pygame.K_SPACE:
                manifold = t.Manifold.collapseRandom(manifold)
                showBoard(speed_ms=0)
                

```
