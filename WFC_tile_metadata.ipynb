```python
import os
import sys
from turtle import width
if "newtiles" in sys.modules.keys():
    del sys.modules["newtiles"]
import newtiles as t
import pygame
from typing import List
import numpy as np

try:
    pygame.quit()
except:
    pass

tileSetupFilePath = os.path.join("tilesetups", "groupTest.tsf")




tsf : t.TileSetupFile = t.TileSetupFile.read_tsf(tileSetupFilePath)
boardSize = max(tsf.width, tsf.height)

screensize = 900
tilesize = screensize / boardSize

pygame.init()
pygame.font.init()
pygame.display.set_caption('Tiles Demo')
screen = pygame.display.set_mode((screensize * tsf.width/tsf.height, screensize))
pygame.event.pump()
pygame.display.update()
pygame.time.wait(100)

manifold = tsf.construct_manifold()
manifold = t.Manifold.PropagateAll(manifold)

def showBoard(speed_ms : int = 100):
    
    screen.fill((100, 100, 100))

    # draw all the tiles
    for x in range(tsf.width):
        for y in range(tsf.height):
            
            manifold.collections[y * tsf.width + x].draw(screen, np.array([x, y]), tilesize)
            pygame.time.wait(speed_ms // (tsf.width * tsf.height))

            if speed_ms > 0:
                pygame.event.pump()
                pygame.display.update()
    
    # # draw dual grid
    # for x in range(0, boardSize):
    #     pygame.draw.line(screen, (100, 100, 200), ((x+0.5) * tilesize, 0), ((x+0.5) * tilesize, screensize), width=5)
    #     pygame.draw.line(screen, (100, 100, 200), (0, (x+0.5) * tilesize), (screensize, (x+0.5) * tilesize), width=5)

        # draw normal grid
    for x in range(1, tsf.width):
        pygame.draw.line(screen, (0, 0, 0), (x * tilesize, 0), (x * tilesize, screensize), width=1) # vertical
    for y in range(1, tsf.height):
        pygame.draw.line(screen, (0, 0, 0), (0, y * tilesize), (screensize, y * tilesize), width=1) # horizonta
    
    pygame.event.pump()
    pygame.display.update()


def showAttr():
    
    screen.fill((100, 100, 100))

    # draw all the tiles
    for x in range(tsf.width):
        for y in range(tsf.height):
            
            manifold.collections[y * tsf.width + x].draw_attrs(screen, np.array([x, y]), tilesize)
    
    pygame.event.pump()
    pygame.display.update()


showBoard(speed_ms=200)

while(not manifold.finished()):
    
    cont = False
    while not cont:
        for event in pygame.event.get():
            if event.type == pygame.QUIT or (event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE):
                pygame.quit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    cont = True
                    break
                if event.key == pygame.K_c:
                    showAttr()
                    break
    
    manifold = t.Manifold.collapseRandom(manifold)
    showBoard(speed_ms=0)



while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT or (event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE):
                pygame.quit()
```
